```python
#Importar librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Creamos una funciÃ³n para no repetir codigo que nos sirva para agrupar y hacer el promedio de la columna que seleccionamos
def agrupar_y_promedio(columna_a_agrupar, columna_a_modificar):
    #Agrupamos y aplicamos la funcion
    agrupado = df.groupby(columna_a_agrupar)[columna_a_modificar].mean().reset_index()
    #Ordenamos los valores por la columna a la que le aplicamos la funcion y reseteamos los indices
    agrupado = agrupado.sort_values(by=columna_a_modificar).reset_index(drop=True)
    return agrupado
```

```python
#Aplicamos la funciÃ³n para agrupar por tamaÃ±o de motor y cuantos kw consume en 100km
km_ciudad = agrupar_y_promedio("Motor (kW)", "City (kWh/100 km)")
km_ciudad
```

```python
#Hacemos un top10 de los que menos consumen
km_ciudad_top10 = km_ciudad.head(10)
km_ciudad_top10
```

```python
#Hacemos el grafico
sns.barplot(data = km_ciudad_top10, x= "Motor (kW)", y= "City (kWh/100 km)", order=km_ciudad_top10["Motor (kW)"])
plt.show()
```

```python
#Los que mas consumen
km_ciudad_bottom10 = km_ciudad.tail(10)
km_ciudad_bottom10 = km_ciudad_bottom10.sort_values(by="City (kWh/100 km)", ascending=False).reset_index(drop=True)
km_ciudad_bottom10
```

```python
#Hacemos el grafico
sns.barplot(data = km_ciudad_bottom10, x= "Motor (kW)", y= "City (kWh/100 km)", order=km_ciudad_bottom10["Motor (kW)"])
plt.show()
```

```python
#Usamos la funcino para ver que motores son los que pueden recorrer mas km
motor_rango = agrupar_y_promedio("Motor (kW)", "Range (km)")
motor_rango
```

```python
#Hacemos el top10 de los que menos rango tienen y lo graficamos
motor_rango_menoskm = pd.DataFrame(motor_rango.head(10))
sns.barplot(data = motor_rango_menoskm, x= "Motor (kW)", y= "Range (km)", order=motor_rango_menoskm["Motor (kW)"])
plt.show()
```

```python
#Hacemos el top10 de los que mas rango tienen y lo graficamos
motor_rango_maskm = pd.DataFrame(motor_rango.tail(10))
motor_rango_maskm = motor_rango_maskm.sort_values(by="Range (km)", ascending=False).reset_index(drop=True)
sns.barplot(data = motor_rango_maskm, x= "Motor (kW)", y= "Range (km)", order=motor_rango_maskm["Motor (kW)"])
plt.show()
```

```python
#Los 10 motores que menos tardan en cargarse y lo graficamos
promedio_carga_menorcarga = promedio_carga.head(10)
sns.barplot(data = promedio_carga_menorcarga, x= "Motor (kW)", y= "Recharge time (h)", order=promedio_carga_menorcarga["Motor (kW)"])
plt.show()
```

```python
#Los 10 motores que mas tardan en cargarse y lo graficamos
promedio_carga_mayorcarga = promedio_carga.tail(10)
promedio_carga_mayorcarga = promedio_carga_mayorcarga.sort_values(by="Recharge time (h)", ascending=False).reset_index(drop=True)
sns.barplot(data = promedio_carga_mayorcarga, x= "Motor (kW)", y= "Recharge time (h)", order=promedio_carga_mayorcarga["Motor (kW)"])
plt.show()
```

```python
#Filtramos por el modelo del auto
hyundai_ioniq5 = df[(df["Make"] == "Hyundai") & (df["Model"].str.contains("IONIQ 5", case=False, na=False) & (df["Model year"] == 2023))]
hyundai_ioniq5
```

```python
#Filtramos por el modelo del auto
kia_niro = df[(df["Make"] == "Kia") & (df["Model"].str.contains("Niro", case=False, na=False) & (df["Model year"] > 2021) & (df["Model year"] < 2024))]
kia_niro
```

```python
#Filtramos por el modelo del auto
tesla_model3 = df[(df["Make"] == "Tesla") & (df["Model"].str.contains("Model 3 Long|RWD|AWD", case=False, na=False) & (df["Model year"] > 2018) & (df["Model year"] < 2024))]
tesla_model3.head()
```

```python
#Filtramos por el modelo del auto
toyota_bz4x = df[(df["Make"] == "Toyota") & (df["Model"].str.contains("bz4x", case=False, na=False) & (df["Model year"] == 2023))]
toyota_bz4x
```

```python
#Unimos todos los datos filtrados
autos_permitidos = pd.concat([hyundai_ioniq5, kia_niro, tesla_model3, toyota_bz4x]).reset_index(drop=True)
autos_permitidos["Marca_aÃ±o_y_modelo"] = autos_permitidos["Model year"].astype(str) + " " + autos_permitidos["Make"].astype(str) + " " + autos_permitidos["Model"].astype(str)
```

```python
#Graficamos los modelos con su tiempo de carga
plt.figure(figsize=(10,8))
sns.barplot(data= autos_permitidos, y = "Marca_aÃ±o_y_modelo", x= "Recharge time (h)")
plt.show()
```

```python
#Graficamos los modelos con los kwh que consumen en 100km
plt.figure(figsize=(10,8))
sns.barplot(data= autos_permitidos, y = "Marca_aÃ±o_y_modelo", x= "City (kWh/100 km)")
plt.show()
```

```python
#Graficamos los modelos con la distancia que pueden recorrer
plt.figure(figsize=(10,8))
sns.barplot(data= autos_permitidos, y = "Marca_aÃ±o_y_modelo", x= "Range (km)")
plt.show()
```
