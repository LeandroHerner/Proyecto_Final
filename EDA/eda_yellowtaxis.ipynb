```python
#Importamos las librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

No hay una gran cantidad de duplicados

La fecha y la hora está todo junto, con lo que voy a separarlas para poder hacer un mejor analisis

```python
#Pasamos la columna a datetime
df["tpep_pickup_datetime"] = pd.to_datetime(df["tpep_pickup_datetime"])

#Creamos nuevas columnas para el año, mes, día y hora
df["year"] = df["tpep_pickup_datetime"].dt.year
df["month"] = df["tpep_pickup_datetime"].dt.month
df["pickup_day"] = df["tpep_pickup_datetime"].dt.day
df["pickup_hour"] = df["tpep_pickup_datetime"].dt.hour
df["pickup_minute"] = df["tpep_pickup_datetime"].dt.minute

df["tpep_dropoff_datetime"] = pd.to_datetime(df["tpep_dropoff_datetime"])

df["dropoff_day"] = df["tpep_dropoff_datetime"].dt.day
df["dropoff_hour"] = df["tpep_dropoff_datetime"].dt.hour
df["dropoff_minute"] = df["tpep_dropoff_datetime"].dt.minute
```

```python
df
```

```python
#Agrupamos por la hora en la que se inicio el viaje y contamos la cantidad de registros
franja_horaria = df.groupby("pickup_hour")["VendorID"].count().reset_index()
#Cambiamos de nombre la columna a cantidad_viajes
franja_horaria.rename(columns={"VendorID": "cantidad_viajes"}, inplace=True)
```

```python
franja_horaria
```

```python
#Graficamos
sns.lineplot(data = franja_horaria, x="pickup_hour", y="cantidad_viajes", marker="o")
plt.xticks(range(min(franja_horaria["pickup_hour"]), max(franja_horaria["pickup_hour"])+1, 1))
plt.show()
```

```python
#Agrupamos por año y contamos los registros
anio = df.groupby("year")["pickup_hour"].count().reset_index()
#Renombramos la columna a cantidad_viajes
anio.rename(columns={"pickup_hour" : "cantidad_viajes"}, inplace=True)
```

```python
anio
```

Vemos que hay registros de años que no corresponden con el analisis, por lo que voy a eliminar las filas que no correspondan

```python
#Eliminamos los años que no vamos a utilizar
anio.drop(anio[(anio["year"] > 2001) & (anio["year"] < 2019)].index, inplace=True)
#Reseteamos los indices
anio = anio.reset_index(drop=True)
#Cambiamos el tipo de columna a entero
anio["cantidad_viajes"] = anio["cantidad_viajes"].astype(int)
anio
```

```python
#Graficamos
sns.lineplot(data = anio, x="year", y="cantidad_viajes", marker="o")
plt.xticks(range(min(anio["year"]), max(anio["year"])+1, 1))
plt.ticklabel_format(style="plain", axis="y")
plt.show()
```

```python
#Agrupamos por mes y contamos la cantidad de registros
mes = df.groupby("month")["VendorID"].count().reset_index()
#Renombramos la columna
mes.rename(columns={"VendorID" : "cantidad_viajes"}, inplace=True)
mes
```

```python
#Graficamos
sns.lineplot(data = mes, x="month", y="cantidad_viajes", marker="o")
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(range(min(mes["month"]), max(mes["month"])+1, 1))
plt.show()
```

```python
#Agrupamos con el metodo de pago y contamos los registros
metodo_pago = df.groupby("payment_type")["VendorID"].count().reset_index()
#Renombramos la columna
metodo_pago.rename(columns={"VendorID" : "Cantidad_metodo_pago"}, inplace=True)
metodo_pago
```

```python
#Graficamos
sns.barplot(data= metodo_pago, x="payment_type", y="Cantidad_metodo_pago")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
#Agrupamos por donde empieza el viaje y contamos los registros
locacion_inicio = df.groupby("PULocationID")["VendorID"].count().reset_index()
#Renombramos la columna
locacion_inicio.rename(columns={"VendorID" : "Cantidad_locacion"}, inplace=True)
#Ordenamos los valores
locacion_inicio = locacion_inicio.sort_values(by="Cantidad_locacion", ascending=False).reset_index(drop=True)
locacion_inicio
```

```python
#Graficamos el top10 de ubicaciones en las que se inician viajes
locacion_inicio_top10 = locacion_inicio.head(10)
sns.barplot(data = locacion_inicio_top10, x= "PULocationID", y="Cantidad_locacion", order=locacion_inicio_top10["PULocationID"], color="red")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
#Agrupamos por destino y contamos los registros
locacion_fin = df.groupby("DOLocationID")["VendorID"].count().reset_index()
#Renombramos la columna
locacion_fin.rename(columns={"VendorID" : "Cantidad_locacion"}, inplace=True)
#Ordenamos los valores
locacion_fin = locacion_fin.sort_values(by="Cantidad_locacion", ascending=False).reset_index(drop=True)
locacion_fin
```

```python
#Graficamos el top10 de ubicaciones en las que se terminan los viajes
locacion_fin_top10 = locacion_fin.head(10)
sns.barplot(data = locacion_fin_top10, x= "DOLocationID", y="Cantidad_locacion", order=locacion_fin_top10["DOLocationID"], color="red")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
#Agrupamos por mes y hacemos el promedio de cuanto se cobran los viajes
recaudacion_promedio_por_viaje_mes = df.groupby("month")["fare_amount"].mean().reset_index()
recaudacion_promedio_por_viaje_mes
#Graficamos
sns.barplot(data = recaudacion_promedio_por_viaje_mes, x="month", y="fare_amount")
plt.show()
```

```python
#Agrupamos por mes y sumamos el total de la recaudacion
recaudacion_total_por_mes = df.groupby("month")["fare_amount"].sum().reset_index()
#Graficamos
sns.barplot(data = recaudacion_total_por_mes, x = "month", y="fare_amount")
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
