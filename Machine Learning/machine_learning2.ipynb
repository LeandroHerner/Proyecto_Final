```python
df_machinelearning = taxis[["PULocationID", "date", "temperature_2m", "precipitation_mm"]]

dias_a_numeros = {
    'Monday': 1,
    'Tuesday': 2,
    'Wednesday': 3,
    'Thursday': 4,
    'Friday': 5,
    'Saturday': 6,
    'Sunday': 7
}

df_machinelearning["date"] = pd.to_datetime(df_machinelearning["date"])
df_machinelearning['dia'] = df_machinelearning['date'].dt.day_name().map(dias_a_numeros)
df_machinelearning["mes"] = df_machinelearning["date"].dt.month
df_machinelearning["hora"] = df_machinelearning["date"].dt.hour

df_machinelearning = df_machinelearning.reset_index(drop=True)

groupby = df_machinelearning.groupby(["PULocationID", "dia", "mes", "hora", "temperature_2m", "precipitation_mm"])["date"].count().reset_index(name="demand")
groupby
```

```python
# Calcula los cuartiles para determinar los límites
cuartiles = groupby['demand'].quantile([0.25, 0.5, 0.75])

# Define los límites basados en los cuartiles
limite_muy_baja = cuartiles[0.25]
limite_baja = cuartiles[0.5]
limite_normal = cuartiles[0.75]
limite_alta = groupby['demand'].max()  # Suponiendo que "alta" es superior al tercer cuartil

# Función para categorizar la demanda
def categorizar_demand(demand):
    if demand <= limite_muy_baja:
        return 0
    elif demand <= limite_baja:
        return 1
    elif demand <= limite_normal:
        return 2
    elif demand <= limite_alta:
        return 3

# Aplica la función para categorizar la demanda y crear una nueva columna
groupby['demand'] = groupby['demand'].apply(categorizar_demand)
```

Arboles de decision

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

X = groupby[["PULocationID", "temperature_2m", "precipitation_mm", "dia", "mes", "hora"]]
Y = groupby["demand"]

# Dividir el conjunto de datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=123)

# Inicializar y entrenar el modelo de árbol de decisión para clasificación
decision_tree_classifier = DecisionTreeClassifier(max_depth=18, min_samples_split=6, min_samples_leaf=4, random_state=123)
decision_tree_classifier.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = decision_tree_classifier.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print("Precisión del modelo de árbol de decisión:", accuracy)

#Analizar desde la desviacion estandar, sub dividir los cuartiles, cuanto influye el clima en el modelo de ml
# Sacar los borough que son unkown
#Predecir por mes y borough para el dashboard
```
